```python
import os
import numpy as np
from ultralytics import YOLO
from sklearn.metrics import f1_score
from google.colab import files
from PIL import Image


# # 1. Modeli yÃ¼kle
#model = YOLO("/content/runs/detect/yolo11_model/weights/best.pt")
# 2. Test gÃ¶rÃ¼ntÃ¼lerine tahmin yap
results = model.predict(
    source="/content/drive/Othercomputers/DizuÌˆstuÌˆ BilgisayarÄ±m/MasauÌˆstuÌˆ/veri seti/Human Bone Fracture/Human Bone Fractures Multi-modal Image Dataset (HBFMID)/Bone Fractures Detection/test/images",  # test gÃ¶rÃ¼ntÃ¼lerinin olduÄŸu klasÃ¶r
    save=True,                      # sonuÃ§ gÃ¶rselleri kaydedilsin
    conf=0.25 ,                      # gÃ¼ven skoru eÅŸiÄŸi
    iou=0.6
)

# 3. IoU hesaplama fonksiyonu
def compute_iou(box1, box2):
    x1, y1, x2, y2 = box1
    x1g, y1g, x2g, y2g = box2
    xi1 = max(x1, x1g)
    yi1 = max(y1, y1g)
    xi2 = min(x2, x2g)
    yi2 = min(y2, y2g)
    inter_area = max(0, xi2 - xi1) * max(0, yi2 - yi1)
    union_area = (x2 - x1)*(y2 - y1) + (x2g - x1g)*(y2g - y1g) - inter_area
    return inter_area / union_area if union_area > 0 else 0

# 4. DeÄŸerlendirme
ious = []
y_true = []
y_pred = []

for result in results:
    pred_boxes = result.boxes.xyxy.cpu().numpy()  # [x1, y1, x2, y2]
    img_path = result.path

    # GerÃ§ek label dosyasÄ±nÄ± bul
    txt_path = img_path.replace("images", "labels").replace(".jpg", ".txt")
    if not os.path.exists(txt_path):
        continue

    gt_boxes = []
    with open(txt_path, "r") as f:
        for line in f.readlines():
            cls, x_center, y_center, w, h = map(float, line.strip().split())
            x1 = (x_center - w / 2) * 640
            y1 = (y_center - h / 2) * 640
            x2 = (x_center + w / 2) * 640
            y2 = (y_center + h / 2) * 640
            gt_boxes.append([x1, y1, x2, y2])

    for gt in gt_boxes:
        best_iou = 0
        for pred in pred_boxes:
            iou = compute_iou(gt, pred)
            best_iou = max(best_iou, iou)
        ious.append(best_iou)
        y_true.append(1)
        y_pred.append(1 if best_iou >= 0.2 else 0)

# 5. Ortalama skorlar
mean_iou = np.mean(ious)
f1 = f1_score(y_true, y_pred)

print(f"âœ… Ortalama IoU: {mean_iou:.4f}")
print(f"âœ… F1-Score: {f1:.4f}")

```

```python
import matplotlib.pyplot as plt
import cv2
import os

# GÃ¶rsellerin bulunduÄŸu klasÃ¶rÃ¼ tanÄ±mla
predict_folder = results[0].save_dir

# Dosya isimlerini al (sadece .jpg)
image_files = [f for f in os.listdir(predict_folder) if f.endswith(".jpg")]

# Ä°lk 10 tahmini gÃ¶rselleÅŸtirelim
num_images = min(10, len(image_files))  # klasÃ¶rde 10'dan az varsa uyumlu hale getir
rows, cols = 2, 5

fig, axs = plt.subplots(rows, cols, figsize=(cols * 3, rows * 3))

for i in range(num_images):
    img_path = os.path.join(predict_folder, image_files[i])
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    row = i // cols
    col = i % cols
    axs[row, col].imshow(img)
    axs[row, col].set_title(f"{image_files[i][:15]}...", fontsize=8)
    axs[row, col].axis('off')

# BoÅŸ kalan hÃ¼creleri gizle (eÄŸer 10'dan az gÃ¶rsel varsa)
for j in range(num_images, rows * cols):
    row = j // cols
    col = j % cols
    axs[row, col].axis('off')

plt.tight_layout()
plt.show()
```
